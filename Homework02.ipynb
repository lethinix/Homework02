```python
lowest_wine = []
wine_loc = []

for wine in wine_data[1:]:
    if wine[1] == "US":
        lowest_wine.append(float(wine[3]))
        wine_loc.append(wine[8])

us_min_rating = min(lowest_wine)
lowest_wine_loc = wine_loc[lowest_wine.index(us_min_rating)]

print(us_min_rating)
print(lowest_wine_loc)
```

- Where's the worse wine in the US from ?
Cooper Vineyards


```python
us_sum = spain_sum = france_sum = 0
us_count = spain_count = france_count = 0

for wine in wine_data[1:]:
    price = float(wine[4])
    country = wine[1]

    if country == "US":
        us_sum += price
        us_count += 1
    elif country == "Spain":
        spain_sum += price
        spain_count += 1
    elif country == "France":
        france_sum += price
        france_count += 1

avg_price_us = us_sum / us_count
avg_price_spain = spain_sum / spain_count
avg_price_france = france_sum / france_count

print(avg_price_us)
print(avg_price_spain)
print(avg_price_france)
```

- Which country between spain/us/france has highest average price ?
spain

```python
us_sum = spain_sum = france_sum = 0
us_count = spain_count = france_count = 0

for wine in wine_data[1:]:
    points = float(wine[3])
    country = wine[1]

    if country == "US":
        us_sum += points
        us_count += 1
    elif country == "Spain":
        spain_sum += points
        spain_count += 1
    elif country == "France":
        france_sum += points
        france_count += 1

avg_points_us = us_sum / us_count
avg_points_spain = spain_sum / spain_count
avg_points_france = france_sum / france_count

print(avg_points_us)
print(avg_points_spain)
print(avg_points_france)
```

- Which country between spain/us/france has highest average quality points ?
france

```python
high_sum = low_sum = 0
high_count = low_count = 0

for wine in wine_data[1:]:
    points = float(wine[3])
    price = float(wine[4])

    if points >= 90:
        high_sum += price
        high_count += 1
    else:
        low_sum += price
        low_count += 1

avg_high_price = high_sum / high_count
avg_low_price = low_sum / low_count

print("avg price (90+ points):", avg_high_price)
print("avg price (<90 points):", avg_low_price)
```
